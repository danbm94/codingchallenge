# IMDB Movies review

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
```

```python
movies= pd.read_csv("NetflixOriginals.csv",sep=",", encoding='ISO 8859-1')
```

## Analysis and data visualization

```python
movies.describe(include='all')
```

```python
sns.histplot(movies['Runtime'],bins=25)
plt.title("Distribution of the runtime of the featured films in minutes")
plt.show()

sns.histplot(movies['IMDB Score'],bins=25)
plt.title("Distribution of the IMDB score")
plt.show()
```

```python
movies['Language'].value_counts(ascending=True).plot.barh(figsize=(15,10))
```

```python
df_english = movies[movies['Language'] == 'English'].sort_values(by='IMDB Score', ascending=False)
df_english.head(10)
```

## Answering the questions:
- There are clearly a lot of films that have English as their primary language. Of the 503 films, what proportion have English as their primary language? 

- What is the ratio of English-language films to films in a single language that is NOT English?

- What proportion of the films has multiple primary languages? 


```python
prop= movies.value_counts('Language', ascending=True)
n = len(movies['Language'])

prop['proportion'] = prop/n
prop['percentage'] = prop['proportion']*100
```

```python
prop['percentage'].plot.barh(figsize=(15,10))
```
